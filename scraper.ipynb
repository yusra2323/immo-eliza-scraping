```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import time

def scrape_page(url):
    # Construct the URL for the current page
    url = f'https://www.immoweb.be/en/classified/house/for-sale/trooz/4870/20234194'
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find all property listings on the page
    property_listings = soup.find_all('article', class_='search-results__item')
    
    properties = []
    
    # Loop through each property on the page and extract data
    for listing in property_listings:
    
            
            PropertyID = listing.find('div', {'class': 'classified__header--immoweb-code'}).text.strip()
            Localityname = listing.find('div', {'class': 'classified__information--address','span':'aria-hidden'}).text.strip()
            Postalcode = listing.find('span', {'class': 'classified__information--address-row'}).text.strip()
            Price=listing.find('span', {'class': 'sr-only'}).text.strip().replace("€", "").replace(",", "")
            roomsno = listing.find('p', {'class': 'classified__information--property'}).text.strip()
            living_area = listing.find('div', {'class': 'living-area'}).text.strip()
            Equippedkitchen = listing.find('td', {'class': 'classified-table__data'}).text.strip()
            Terrace = listing.find('div', {'class': 'classified-table__data'}).text.strip()
            Garden = listing.find('td', {'class': 'classified-table__data'}).text.strip()
            Swimmingpool = listing.find('td', {'class': 'classified-table__data'}).text.strip()

     
           
           
           
            # Store the data in a dictionary
            property_data = {
                'PropertyID': PropertyID,
                ' Localityname': Localityname,
                'Postalcode': Postalcode,
                'Price': Price,
                'roomsno': roomsno,
                'living_area': living_area,
                ' Equippedkitchen':  Equippedkitchen,
                'Terrace': Terrace,
                'Garden':Garden,
                'Swimmingpool':Swimmingpool,
  
                
            }
            properties.append(property_data)
        
    return properties

# Test the function with a single property URL



# Loop through multiple pages
all_properties = []
for page_no in range(1, 10):  # Example for first 100 pages
    print(f"Scraping page {page_no}")
    properties = scrape_page(page_no)
    all_properties.extend(properties)
    
    # Be kind to the website, wait between requests
    time.sleep(1)

# Convert to DataFrame and save to CSV
df = pd.DataFrame(all_properties)
df.to_csv('real_estate_data.csv', index=False)
```
